```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(load_iris().data, columns=load_iris().feature_names)
df['target'] = load_iris().target
df = df[df['target'] != 2]
df['target'].value_counts()
```

```python
plt.figure()
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['sepal length (cm)', 'sepal width (cm)']], df['target'], random_state=0)

# Create the SVM classifier
svm_classifier = SVC(kernel='linear', C=1.0)

# Fit the classifier to the training data
svm_classifier.fit(X_train, y_train)

# Plot the margin
plt.figure()
plt.scatter(X_train['sepal length (cm)'], X_train['sepal width (cm)'], c=y_train)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

# Create a meshgrid to plot the decision boundary
x_min, x_max = X_train['sepal length (cm)'].min() - 1, X_train['sepal length (cm)'].max() + 1
y_min, y_max = X_train['sepal width (cm)'].min() - 1, X_train['sepal width (cm)'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and margin
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(svm_classifier.support_vectors_[:, 0], svm_classifier.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k')
plt.title('SVM Classifier')
plt.show()
```
